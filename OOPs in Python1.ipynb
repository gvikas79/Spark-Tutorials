## Object-Oriented Programming (OOP) in Python

Object-Oriented Programming (OOP) is a programming paradigm based on the concept of 'objects', which can contain data (attributes) and code (methods). The primary goal of OOP is to increase the flexibility and maintainability of programs. Python is a multi-paradigm language that fully supports OOP.

### Why OOP?

*   **Modularity**: Objects create a modular structure, making code easier to manage and debug.
*   **Reusability**: Inheritance allows for the reuse of code across different classes.
*   **Flexibility (Polymorphism)**: Objects can take on many forms, allowing for more generic code.
*   **Security (Encapsulation)**: Data can be hidden within an object, protecting it from external interference.
*   **Scalability**: Well-designed OOP systems are easier to extend and maintain as they grow.

### 1. Classes and Objects

*   **Class**: A class is a blueprint or a template for creating objects. It defines a set of attributes (data) and methods (functions) that the created objects will have. Think of it as a cookie cutter.
*   **Object**: An object is an instance of a class. It's a concrete entity created from a class. Think of it as a cookie created from the cutter.

In Python, you define a class using the `class` keyword. The `__init__` method is a special method called the constructor, which is automatically called when a new object is created. `self` refers to the instance of the class itself.

```python
class Dog:
    # Class attribute (shared by all instances)
    species = "Canis familiaris"

    # Constructor method
    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute

    # Instance method
    def bark(self):
        return f"{self.name} says Woof!"

    # Another instance method
    def description(self):
        return f"{self.name} is {self.age} years old."

# Creating objects (instances of the Dog class)
my_dog = Dog("Buddy", 3)
your_dog = Dog("Lucy", 5)

# Accessing attributes
print(f"My dog's name: {my_dog.name}")
print(f"Your dog's age: {your_dog.age}")
print(f"My dog's species: {my_dog.species}")

# Calling methods
print(my_dog.bark())
print(your_dog.description())
```

### 2. Encapsulation

Encapsulation is the bundling of data (attributes) and methods (functions) that operate on the data into a single unit (class). It restricts direct access to some of an object's components, which means that the internal state of an object is protected from external tampering. In Python, encapsulation is achieved through conventions rather than strict keywords.

*   **Public attributes/methods**: Can be accessed from anywhere.
*   **Protected attributes/methods**: Denoted by a single leading underscore (`_name`). They are intended for internal use within the class or by its subclasses, but can still be accessed externally (a convention, not enforced).
*   **Private attributes/methods**: Denoted by double leading underscores (`__name`). Python performs 'name mangling' to make them harder to access directly from outside the class, but they are not truly private.


```python
class BankAccount:
    def __init__(self, account_holder, initial_balance):
        self.account_holder = account_holder  # Public attribute
        self._account_number = "123456789"    # Protected attribute (by convention)
        self.__balance = initial_balance      # Private attribute (name mangled)

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def get_balance(self):
        return self.__balance

# Create an account
account = BankAccount("Alice Smith", 1000)

# Accessing public attribute
print(f"Account holder: {account.account_holder}")

# Attempting to access protected attribute (by convention, it's discouraged but possible)
print(f"Account number (protected): {account._account_number}")

# Attempting to access private attribute (will cause an AttributeError without name mangling)
# print(account.__balance) # This would raise an AttributeError

# Accessing private attribute via name mangling (not recommended for direct use)
print(f"Balance (via name mangling): {account._BankAccount__balance}")

# Using public methods to interact with the private balance
print(f"Current balance: ${account.get_balance()}")
account.deposit(500)
account.withdraw(200)
print(f"Final balance: ${account.get_balance()}")
account._account_number = 4323242
account.__balance = 4223
```

### 3. Inheritance

Inheritance is a mechanism that allows a new class (subclass/child class) to inherit properties and behavior (attributes and methods) from an existing class (superclass/parent class). This promotes code reusability and establishes a natural 'is-a' relationship (e.g., a 'Car' *is a* 'Vehicle').

*   **Parent Class (Base Class)**: The class being inherited from.
*   **Child Class (Derived Class)**: The class that inherits from the parent class.

Python supports single inheritance (a class inherits from one parent) and multiple inheritance (a class inherits from multiple parents).


```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"Brand: {self.brand}, Model: {self.model}"

    def start_engine(self):
        return "Engine started."

class Car(Vehicle):
    def __init__(self, brand, model, num_doors):
        super().__init__(brand, model) # Call the parent class's constructor
        self.num_doors = num_doors

    # Override parent method
    def display_info(self):
        return f"Car - {super().display_info()}, Doors: {self.num_doors}"

    def honk(self):
        return "Honk! Honk!"

class Motorcycle(Vehicle):
    def __init__(self, brand, model, top_speed):
        super().__init__(brand, model)
        self.top_speed = top_speed

    # Add a new method specific to Motorcycle
    def wheelie(self):
        return "Doing a wheelie!"

# Create objects
my_car = Car("Toyota", "Camry", 4)
my_motorcycle = Motorcycle("Harley-Davidson", "Iron 883", 110)

# Accessing inherited attributes and methods
print(my_car.display_info())
print(my_car.start_engine())
print(my_car.honk())

print(my_motorcycle.display_info())
print(my_motorcycle.start_engine())
print(my_motorcycle.wheelie())
```

### 4. Polymorphism

Polymorphism means 'many forms'. In OOP, it refers to the ability of different classes to be treated as instances of a common type, and for methods with the same name to behave differently based on the object they are called on. It allows you to write more generic and flexible code.

Polymorphism can be achieved through:

*   **Method Overriding**: A child class provides its own implementation of a method that is already defined in its parent class (as seen in the `display_info` example above).
*   **Method Overloading**: Defining multiple methods with the same name but different parameters (Python doesn't support traditional method overloading directly, but can be simulated using default arguments or `*args`/`**kwargs`).
*   **Duck Typing**: If it walks like a duck and quacks like a duck, then it's a duck. Python's dynamic typing allows objects of different types to be used interchangeably if they support the same method calls.


```python
class Bird:
    def speak(self):
        return "Generic bird sound"

class Sparrow(Bird):
    def speak(self):
        return "Chirp, chirp!"

class Parrot(Bird):
    def speak(self):
        return "Squawk! Hello!"

class Dog:
    # A non-Bird class that also has a 'speak' method
    def speak(self):
        return "Woof! Woof!"

# A function that expects an object with a 'speak' method
def make_animal_speak(animal):
    print(animal.speak())

# Demonstrating polymorphism
bird = Bird()
sparrow = Sparrow()
parrot = Parrot()
another_dog = Dog() # Even though Dog is not related by inheritance, it works due to duck typing

make_animal_speak(bird)
make_animal_speak(sparrow)
make_animal_speak(parrot)
make_animal_speak(another_dog)

# Example of method overriding from previous example:
my_car = Car("Ford", "Focus", 5)
my_vehicle = Vehicle("Boeing", "747")

print(my_car.display_info())      # Car's overridden method
print(my_vehicle.display_info())  # Vehicle's original method
```

### 5. Abstraction

Abstraction means showing only essential features of an object and hiding the internal implementation details. It focuses on 'what' an object does rather than 'how' it does it. In Python, abstraction can be achieved using Abstract Base Classes (ABCs) from the `abc` module. An abstract class cannot be instantiated directly and often contains one or more abstract methods (methods declared but not implemented).


```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    def description(self):
        return "This is a generic shape."

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14159 * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

# You cannot instantiate an abstract class directly
# shape = Shape() # This would raise a TypeError

circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Circle Area: {circle.area()}")
print(f"Circle Perimeter: {circle.perimeter()}")
print(circle.description())

print(f"Rectangle Area: {rectangle.area()}")
print(f"Rectangle Perimeter: {rectangle.perimeter()}")
print(rectangle.description())
```

```python
# --- Solutions for Advanced-Level Practice Questions ---

import datetime
import uuid
from abc import ABC, abstractmethod

print("\n--- Question 11 Solutions ---")

# 1. Subject Abstract Base Class
class Subject(ABC):
    def __init__(self):
        self._observers = []

    def attach(self, observer):
        if observer not in self._observers:
            self._observers.append(observer)

    def detach(self, observer):
        try:
            self._observers.remove(observer)
        except ValueError:
            pass # Observer not found

    def notify_observers(self, data):
        for observer in self._observers:
            observer.update(data)

# 2. Concrete Subject: StockMarket
class StockMarket(Subject):
    def __init__(self, stock_name, initial_price):
        super().__init__()
        self.stock_name = stock_name
        self._stock_price = initial_price
        print(f"StockMarket '{self.stock_name}' initialized with price: ${self._stock_price:.2f}")

    def set_price(self, new_price):
        if new_price != self._stock_price:
            old_price = self._stock_price
            self._stock_price = new_price
            print(f"\n--- {self.stock_name} price changed from ${old_price:.2f} to ${new_price:.2f} ---")
            self.notify_observers({'stock': self.stock_name, 'price': new_price, 'timestamp': datetime.datetime.now()})
        else:
            print(f"\n{self.stock_name} price remains ${new_price:.2f}")

# 3. Observer Abstract Base Class
class Observer(ABC):
    @abstractmethod
    def update(self, data):
        pass

# 4. Concrete Observers: Trader and Analyst
class Trader(Observer):
    def __init__(self, name, buy_threshold, sell_threshold):
        self.name = name
        self.buy_threshold = buy_threshold
        self.sell_threshold = sell_threshold

    def update(self, data):
        stock = data['stock']
        price = data['price']
        if price <= self.buy_threshold:
            print(f"Trader {self.name}: {stock} price (${price:.2f}) is low. BUY signal!")
        elif price >= self.sell_threshold:
            print(f"Trader {self.name}: {stock} price (${price:.2f}) is high. SELL signal!")
        else:
            print(f"Trader {self.name}: {stock} price (${price:.2f}) is within range.")

class Analyst(Observer):
    def __init__(self, name):
        self.name = name

    def update(self, data):
        stock = data['stock']
        price = data['price']
        timestamp = data['timestamp'].strftime('%Y-%m-%d %H:%M:%S')
        print(f"Analyst {self.name}: {stock} updated to ${price:.2f} at {timestamp}")

# Demonstration of Question 11
apple_stock = StockMarket("AAPL", 150.00)

trader_john = Trader("John", 145.00, 155.00)
analyst_jane = Analyst("Jane")
trader_sara = Trader("Sara", 148.00, 160.00)

apple_stock.attach(trader_john)
apple_stock.attach(analyst_jane)
apple_stock.attach(trader_sara)

apple_stock.set_price(152.00)
apple_stock.set_price(140.00) # Below John's and Sara's buy threshold
apple_stock.set_price(158.00) # Above John's sell threshold, within Sara's
apple_stock.set_price(165.00) # Above both sell thresholds
apple_stock.detach(trader_john)
apple_stock.set_price(147.00)


print("\n--- Question 12 Solutions ---")

# Re-using (or adapting) ProductEncapsulated from Q6
class Product:
    def __init__(self, product_id, name, price, stock):
        self.product_id = product_id
        self.name = name
        self._price = 0 # Initialize with dummy value to use setter
        self._stock = 0 # Initialize with dummy value to use setter
        self.price = price
        self.stock = stock

    @property
    def price(self):
        return self._price

    @price.setter
    def price(self, value):
        if not isinstance(value, (int, float)) or value < 0:
            raise ValueError("Price must be a non-negative number.")
        self._price = value

    @property
    def stock(self):
        return self._stock

    @stock.setter
    def stock(self, value):
        if not isinstance(value, int) or value < 0:
            raise ValueError("Stock must be a non-negative integer.")
        self._stock = value

    def get_available_quantity(self):
        return self.stock

    def decrease_stock(self, quantity):
        if quantity > self.stock:
            raise ValueError(f"Insufficient stock for {self.name}. Available: {self.stock}, Requested: {quantity}")
        self.stock -= quantity
        return True

    def increase_stock(self, quantity):
        if quantity < 0: raise ValueError("Quantity to increase must be positive.")
        self.stock += quantity
        return True

    def __str__(self):
        return f"Product(ID: {self.product_id}, Name: {self.name}, Price: ${self.price:.2f}, Stock: {self.stock})"

# 2. Customer Class
class Customer:
    def __init__(self, name, email):
        self.customer_id = str(uuid.uuid4())
        self.name = name
        self.email = email
        self.past_orders = [] # Stores Order objects

    def __str__(self):
        return f"Customer(ID: {self.customer_id[:8]}..., Name: {self.name}, Email: {self.email})"

# 3. OrderItem Class
class OrderItem:
    def __init__(self, product: Product, quantity: int):
        if not isinstance(product, Product): raise TypeError("product must be a Product instance.")
        if not isinstance(quantity, int) or quantity <= 0: raise ValueError("Quantity must be a positive integer.")
        self.product = product
        self.quantity = quantity

    def get_total_price(self):
        return self.product.price * self.quantity

    def __str__(self):
        return f"  - {self.product.name} (x{self.quantity}) @ ${self.product.price:.2f} each = ${self.get_total_price():.2f}"

# 4. Order Class
class Order:
    STATUS_PENDING = "Pending"
    STATUS_PROCESSING = "Processing"
    STATUS_SHIPPED = "Shipped"
    STATUS_DELIVERED = "Delivered"
    STATUS_CANCELLED = "Cancelled"

    def __init__(self, customer: Customer):
        self.order_id = str(uuid.uuid4())
        self.customer = customer
        self.items = []
        self.order_date = datetime.datetime.now()
        self.status = Order.STATUS_PENDING

    def add_item(self, product: Product, quantity: int):
        # Logic for adding item to order (stock decrement handled by ECommerceSystem)
        order_item = OrderItem(product, quantity)
        self.items.append(order_item)

    def get_total_amount(self):
        return sum(item.get_total_price() for item in self.items)

    def update_status(self, new_status):
        valid_statuses = [self.STATUS_PENDING, self.STATUS_PROCESSING, self.STATUS_SHIPPED, self.STATUS_DELIVERED, self.STATUS_CANCELLED]
        if new_status not in valid_statuses:
            raise ValueError(f"Invalid status: {new_status}. Must be one of {valid_statuses}")
        self.status = new_status
        print(f"Order {self.order_id[:8]}... status updated to {new_status}")

    def __str__(self):
        item_details = "\n".join(str(item) for item in self.items)
        return f"\nOrder ID: {self.order_id[:8]}...\nCustomer: {self.customer.name}\nDate: {self.order_date.strftime('%Y-%m-%d %H:%M:%S')}\nStatus: {self.status}\nItems:\n{item_details}\nTotal: ${self.get_total_amount():.2f}"

# 5. PaymentProcessor Abstract Base Class (re-used from Q8)
class PaymentProcessor(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass

class CreditCardProcessor(PaymentProcessor):
    def process_payment(self, amount):
        print(f"  - Processing credit card payment of ${amount:.2f}...")
        # Simulate payment success/failure
        if amount > 1000:
            print("  - Credit card payment failed (simulated for large amount).")
            return False
        print("  - Credit card payment successful.")
        return True

class PayPalProcessor(PaymentProcessor):
    def process_payment(self, amount):
        print(f"  - Processing PayPal payment of ${amount:.2f}...")
        print("  - PayPal payment successful.")
        return True

# 6. ECommerceSystem Class
class ECommerceSystem:
    def __init__(self):
        self._products = {}
        self._customers = {}
        self._orders = {}

    def add_product(self, product: Product):
        if product.product_id in self._products:
            print(f"Error: Product ID {product.product_id} already exists.")
            return
        self._products[product.product_id] = product
        print(f"Product '{product.name}' added to system.")

    def register_customer(self, name, email):
        customer = Customer(name, email)
        self._customers[customer.customer_id] = customer
        print(f"Customer '{name}' registered with ID {customer.customer_id[:8]}...")
        return customer.customer_id

    def create_order(self, customer_id):
        customer = self._customers.get(customer_id)
        if not customer:
            print(f"Error: Customer with ID {customer_id[:8]}... not found.")
            return None
        order = Order(customer)
        self._orders[order.order_id] = order
        print(f"Order {order.order_id[:8]}... created for {customer.name}.")
        return order.order_id

    def add_product_to_order(self, order_id, product_id, quantity):
        order = self._orders.get(order_id)
        product = self._products.get(product_id)
        if not order: print("Error: Order not found."); return False
        if not product: print("Error: Product not found."); return False
        if order.status != Order.STATUS_PENDING: print(f"Error: Cannot modify order {order_id[:8]}... in '{order.status}' status."); return False

        try:
            product.decrease_stock(quantity)
            order.add_item(product, quantity)
            print(f"  - Added {quantity} x {product.name} to order {order_id[:8]}...")
            return True
        except ValueError as e:
            print(f"  - Failed to add product to order {order_id[:8]}...: {e}")
            return False

    def process_order(self, order_id, payment_method: PaymentProcessor):
        order = self._orders.get(order_id)
        if not order: print("Error: Order not found."); return False
        if not order.items: print("Error: Order is empty."); return False
        if order.status != Order.STATUS_PENDING: print(f"Error: Order {order_id[:8]}... cannot be processed in '{order.status}' status."); return False

        total_amount = order.get_total_amount()
        print(f"Attempting to process payment for Order {order_id[:8]}... (Total: ${total_amount:.2f})")
        if payment_method.process_payment(total_amount):
            order.update_status(Order.STATUS_PROCESSING)
            order.customer.past_orders.append(order) # Add to customer's past orders upon successful processing
            print(f"Order {order_id[:8]}... successfully processed.")
            return True
        else:
            order.update_status(Order.STATUS_CANCELLED)
            print(f"Order {order_id[:8]}... payment failed. Cancelling order and restocking items.")
            # Restock items if payment fails
            for item in order.items:
                self._products[item.product.product_id].increase_stock(item.quantity)
                print(f"  - Restocked {item.quantity} x {item.product.name}.")
            return False

    def ship_order(self, order_id):
        order = self._orders.get(order_id)
        if not order: print("Error: Order not found."); return False
        if order.status != Order.STATUS_PROCESSING: print(f"Error: Cannot ship order {order_id[:8]}... in '{order.status}' status. Must be 'Processing'."); return False
        order.update_status(Order.STATUS_SHIPPED)
        print(f"Order {order_id[:8]}... has been shipped.")
        return True

    def deliver_order(self, order_id):
        order = self._orders.get(order_id)
        if not order: print("Error: Order not found."); return False
        if order.status != Order.STATUS_SHIPPED: print(f"Error: Cannot deliver order {order_id[:8]}... in '{order.status}' status. Must be 'Shipped'."); return False
        order.update_status(Order.STATUS_DELIVERED)
        print(f"Order {order_id[:8]}... has been delivered.")
        return True

    def get_customer_orders(self, customer_id):
        customer = self._customers.get(customer_id)
        if not customer: return []
        return customer.past_orders # Returns list of Order objects

    def display_product_inventory(self):
        print("\n--- Current Product Inventory ---")
        if not self._products: print("No products in inventory."); return
        for product in self._products.values():
            print(product)
        print("-------------------------------")

# Demonstration of Question 12
ecom_system = ECommerceSystem()

# Add products
prod1 = Product("P001", "Wireless Mouse", 25.00, 10)
prod2 = Product("P002", "Mechanical Keyboard", 80.00, 5)
prod3 = Product("P003", "4K Monitor", 350.00, 3)
prod4 = Product("P004", "Webcam", 50.00, 0) # Out of stock

ecom_system.add_product(prod1)
ecom_system.add_product(prod2)
ecom_system.add_product(prod3)
ecom_system.add_product(prod4)

ecom_system.display_product_inventory()

# Register customers
cust_id_alice = ecom_system.register_customer("Alice Smith", "alice@example.com")
cust_id_bob = ecom_system.register_customer("Bob Johnson", "bob@example.com")

# --- Alice's Order (Successful) ---
print("\n--- Alice's Order (Successful) ---")
alice_order_id = ecom_system.create_order(cust_id_alice)
ecom_system.add_product_to_order(alice_order_id, "P001", 2) # Mouse
ecom_system.add_product_to_order(alice_order_id, "P002", 1) # Keyboard

ecom_system.display_product_inventory() # See stock changes

credit_card_processor = CreditCardProcessor()
ecom_system.process_order(alice_order_id, credit_card_processor)

ecom_system.ship_order(alice_order_id)
ecom_system.deliver_order(alice_order_id)
print(ecom_system._orders[alice_order_id])

# --- Bob's Order (Payment Failure) ---
print("\n--- Bob's Order (Payment Failure) ---")
bob_order_id = ecom_system.create_order(cust_id_bob)
ecom_system.add_product_to_order(bob_order_id, "P003", 2) # 4K Monitor (Total $700)
ecom_system.add_product_to_order(bob_order_id, "P001", 1) # Mouse

ecom_system.display_product_inventory() # See stock changes

ecom_system.process_order(bob_order_id, credit_card_processor) # This should fail due to simulated amount limit

ecom_system.display_product_inventory() # Stock should be restocked
print(ecom_system._orders[bob_order_id])

# --- Another Order (Insufficient Stock) ---
print("\n--- Another Order (Insufficient Stock) ---")
alice_order_id_2 = ecom_system.create_order(cust_id_alice)
ecom_system.add_product_to_order(alice_order_id_2, "P002", 50) # Too much keyboard
ecom_system.add_product_to_order(alice_order_id_2, "P004", 1) # Out of stock webcam

ecom_system.display_product_inventory()

# View customer's past orders
print("\n--- Alice's Past Orders ---")
for order in ecom_system.get_customer_orders(cust_id_alice):
    print(order)

print("\n--- Bob's Past Orders ---")
for order in ecom_system.get_customer_orders(cust_id_bob):
    print(order)
```
