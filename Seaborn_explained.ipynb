```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = sns.load_dataset('penguins')
```

```python
data.head()
```

```python
#Scatter Plot
sns.set_style('darkgrid')
sns.set_palette('Set1')
sns.set_context('notebook')
sns.scatterplot(data = data, x= 'flipper_length_mm', y = 'body_mass_g',hue = 'sex')
```

```python
sns.scatterplot(data = data, x= 'species', y = 'body_mass_g', hue = 'island', style = 'sex', alpha = 0.5)
```

```python
sns.stripplot(data = data, x= 'species', y = 'body_mass_g', hue = 'island', dodge=True, jitter=True)
```

```python
sns.set_context('talk')
sns.despine()
sns.swarmplot(data = data, x= 'species', y = 'body_mass_g', hue = 'island')
```

```python
#Histogram
sns.set_context('notebook')
sns.histplot(data = data, x= 'body_mass_g',hue='island')
```

```python
sns.regplot(data= data, x= 'body_mass_g', y = 'flipper_length_mm')
```

```python
sns.set_style('whitegrid')
sns.lineplot(data= data, x= 'body_mass_g', y = 'flipper_length_mm',hue='island',style='sex')
```

```python
#jointplots
sns.jointplot(data= data, x= 'body_mass_g', y = 'flipper_length_mm')
```

```python
sns.jointplot(data= data, x= 'body_mass_g', y = 'flipper_length_mm', hue='sex', kind = 'kde')
```

```python
sns.barplot(data= data, x= 'species', y = 'body_mass_g',hue ='sex',estimator = np.sum)
```

```python
#countplot
sns.countplot(data = data, x = 'species', hue = 'sex')
```

```python
#boxplot

sns.boxplot(data = data, x ='species', y= 'body_mass_g', hue= 'sex')
```

```python
#violinplot

sns.violinplot(data = data, x ='species', y= 'body_mass_g', hue= 'sex',split =True)
```

```python
sns.violinplot(data = data, x ='species', y= 'body_mass_g', palette='pastel')
sns.swarmplot(data = data, x ='species', y= 'body_mass_g', hue= 'sex', color = 'black', size=3, palette='dark')
```

```python
#KDE -> estimating the probability density function of continuous or non-parametric data
#nothing but a smooth curve that shows the distribution of data - like a smoothed-out version
#it helps you to see where the data is concentrated(peaks) and how is the spread

sns.kdeplot(data = data, x= 'body_mass_g', hue ='species', palette = 'pastel', fill=True)
sns.despine()
```

```python
#Heatmap
data.columns
columns = ['bill_length_mm', 'bill_depth_mm','flipper_length_mm', 'body_mass_g']
#data[columns].corr()
sns.heatmap(data = data[columns].corr(), annot=True, cmap = 'Blues', vmin= 0.2, linewidths="2")
```

```python
#Rug Plot

sns.rugplot(data = data, y= 'body_mass_g', hue = 'species', palette= 'Set2',height= 0.2)
```

```python
# PairPlot

sns.pairplot(data = data, hue= 'sex', palette='Set2', diag_kind= 'bar')
```

```python
#pairgrid

graph = sns.PairGrid(data = data, hue= 'sex', palette='Set2')
graph.map_upper(sns.lineplot)
graph.map_lower(sns.scatterplot)
graph.map_diag(sns.histplot)
```
