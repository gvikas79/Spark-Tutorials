```python
import pandas as pd
import numpy as np

np.random.seed(42)

customers = pd.DataFrame({
    "customer_id": range(1001, 1051),
    "customer_name": [f"Customer_{i}" for i in range(1, 51)],
    "age": np.random.randint(21, 65, 50),
    "city": np.random.choice(
        ["Mumbai", "Delhi", "Chennai", "Bangalore", "Hyderabad"],
        50
    ),
    "income": np.random.randint(300000, 2000000, 50),
    "signup_date": pd.to_datetime("2022-01-01") +
                   pd.to_timedelta(np.random.randint(0, 365, 50), unit="D")
})

print(customers.head())

accounts = pd.DataFrame({
    "account_id": range(2001, 2071),
    "customer_id": np.random.choice(customers["customer_id"], 70),
    "account_type": np.random.choice(
        ["Savings", "Current"],
        70,
        p=[0.7, 0.3]
    ),
    "balance": np.random.randint(5000, 500000, 70),
    "open_date": pd.to_datetime("2022-01-01") +
                 pd.to_timedelta(np.random.randint(0, 365, 70), unit="D")
})

print(accounts.head())

transactions = pd.DataFrame({
    "transaction_id": range(5001, 5201),
    "account_id": np.random.choice(accounts["account_id"], 200),
    "transaction_date": pd.to_datetime("2023-01-01") +
                        pd.to_timedelta(np.random.randint(0, 180, 200), unit="D"),
    "transaction_type": np.random.choice(
        ["Credit", "Debit"],
        200
    ),
    "amount": np.random.randint(500, 100000, 200),
    "channel": np.random.choice(
        ["ATM", "Online", "Branch", "UPI"],
        200
    )
})

print(transactions.head())

loans = pd.DataFrame({
    "loan_id": range(9001, 9031),
    "customer_id": np.random.choice(customers["customer_id"], 30),
    "loan_amount": np.random.randint(100000, 1000000, 30),
    "loan_type": np.random.choice(
        ["Home", "Car", "Personal"],
        30
    ),
    "interest_rate": np.random.uniform(7.5, 14.0, 30).round(2),
    "loan_status": np.random.choice(
        ["Active", "Closed", "Defaulted"],
        30,
        p=[0.6, 0.3, 0.1]
    )
})

print(loans.head())
```
