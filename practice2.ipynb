```python
import pandas as pd
import numpy as np

np.random.seed(10)

customers = pd.DataFrame({
    "customer_id": range(1001, 1021),
    "customer_name": [f"Customer_{i}" for i in range(1, 21)],
    "city": np.random.choice(
        ["Mumbai", "Delhi", "Chennai", "Bangalore"],
        20
    ),
    "signup_date": pd.to_datetime("2023-01-01") +
                   pd.to_timedelta(np.random.randint(0, 180, 20), unit="D")
})

print(customers.head())

products = pd.DataFrame({
    "product_id": range(201, 211),
    "product_name": [f"Product_{i}" for i in range(1, 11)],
    "category": np.random.choice(
        ["Electronics", "Clothing", "Groceries"],
        10
    ),
    "price": np.random.randint(500, 20000, 10)
})

print(products.head())

orders = pd.DataFrame({
    "order_id": range(5001, 5101),
    "customer_id": np.random.choice(customers["customer_id"], 100),
    "product_id": np.random.choice(products["product_id"], 100),
    "order_date": pd.to_datetime("2023-06-01") +
                  pd.to_timedelta(np.random.randint(0, 90, 100), unit="D"),
    "quantity": np.random.randint(1, 5, 100)
})

print(orders.head())

payments = pd.DataFrame({
    "payment_id": range(8001, 8091),
    "order_id": np.random.choice(orders["order_id"], 90, replace=False),
    "payment_method": np.random.choice(
        ["UPI", "Card", "Net Banking", "Cash"],
        90
    ),
    "payment_status": np.random.choice(
        ["Completed", "Failed", "Pending"],
        90,
        p=[0.75, 0.15, 0.10]
    )
})

print(payments.head())
```
