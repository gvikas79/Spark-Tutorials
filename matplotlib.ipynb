```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('coffee.csv')
data
```

```python
espresso = data[data['Coffee Type'] == 'Espresso']
latte = data[data['Coffee Type'] == 'Latte']
latte.head()
```

```python
plt.figure(figsize = (6,6))
plt.plot(espresso['Day'], espresso['Units Sold'], linestyle ='-.',linewidth = 1.5, marker = 'o', label = "Espresso")
plt.plot(latte['Day'], latte['Units Sold'], color = 'Green', linestyle ='-.',linewidth = 1.5, marker = 'o', label = "Latte")
plt.title("Daily count of Units Sold for Espresso and Latte")
plt.xlabel('Day of the week')
plt.ylabel('Units Sold per day')
plt.legend()
plt.ylim(0,50)
plt.xticks(['Monday', 'Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ['Mon','Tue','Wed','Thu','Fri','Sat', 'Sun'])
plt.grid(color = 'gray', linestyle =':', linewidth = 0.5)
#plt.xticks(['Mon','Tue', 'Wed', 'Thu','Fri'], [1,2,3,4,5])
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.bar(espresso['Day'], espresso['Units Sold'], color = 'coral', width = 0.5, label = "Espresso")
plt.bar(latte['Day'], latte['Units Sold'], color = 'purple', width = 0.5,label = "Latte")
plt.title("Daily count of Units Sold for Espresso and Latte")
plt.xlabel('Day of the week')
plt.ylabel('Units Sold per day')
plt.legend()
plt.ylim(0,50)
plt.xticks(['Monday', 'Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ['Mon','Tue','Wed','Thu','Fri','Sat', 'Sun'])
plt.grid(color = 'gray', linestyle =':', linewidth = 0.5)
#plt.xticks(['Mon','Tue', 'Wed', 'Thu','Fri'], [1,2,3,4,5])
plt.show()
```

```python
plt.subplot(2,2,1)
plt.bar(espresso['Day'], espresso['Units Sold'], color = 'coral', width = 0.5, label = "Espresso")
plt.title("Daily count of Units Sold for Espresso")
plt.xlabel('Day of the week')
plt.ylabel('Units Sold per day')
plt.legend()
plt.ylim(0,50)
plt.xticks(['Monday', 'Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ['Mon','Tue','Wed','Thu','Fri','Sat', 'Sun'])
plt.grid(color = 'gray', linestyle =':', linewidth = 0.5)
#plt.xticks(['Mon','Tue', 'Wed', 'Thu','Fri'], [1,2,3,4,5])

plt.subplot(2,2,2)
plt.bar(latte['Day'], latte['Units Sold'], color = 'purple', width = 0.5,label = "Latte")
plt.title("Daily count of Units Sold for Latte")
plt.xlabel('Day of the week')
plt.ylabel('Units Sold per day')
plt.legend()
plt.ylim(0,50)
plt.xticks(['Monday', 'Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ['Mon','Tue','Wed','Thu','Fri','Sat', 'Sun'])
plt.grid(color = 'gray', linestyle =':', linewidth = 0.5)
#plt.xticks(['Mon','Tue', 'Wed', 'Thu','Fri'], [1,2,3,4,5])
plt.tight_layout()
plt.show()

plt.subplot(2,2,3)
plt.bar(latte['Day'], latte['Units Sold'], color = 'purple', width = 0.5,label = "Latte")
plt.title("Daily count of Units Sold for Latte")
plt.xlabel('Day of the week')
plt.ylabel('Units Sold per day')
plt.legend()
plt.ylim(0,50)
plt.xticks(['Monday', 'Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ['Mon','Tue','Wed','Thu','Fri','Sat', 'Sun'])
plt.grid(color = 'gray', linestyle =':', linewidth = 0.5)
#plt.xticks(['Mon','Tue', 'Wed', 'Thu','Fri'], [1,2,3,4,5])
plt.tight_layout()
plt.show()


```

```python
fig, ax = plt.subplots(2, 1, figsize= (9,7))

ax[0].plot(espresso['Day'], espresso['Units Sold'])
ax[0].set_title("Daily count of Units Sold for Espresso")
ax[1].bar(latte['Day'], latte['Units Sold'])
ax[1].set_title("Daily count of Units Sold for Latte")
fig.suptitle('Sales Count for Espresso and Latte')

plt.tight_layout()
plt.show()
```
