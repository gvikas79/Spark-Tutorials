```python
data.rdd.getNumPartitions()
```

```python
from pyspark.sql.functions import col, expr
data_filtered = data.select(col("employee_id"), expr("name"), data.age, data.salary).where("salary > 40000")
data_filtered.show()
```

```python
data_casted = data_filtered.select(expr("employee_id as emp_id"), data.name, expr("cast(age as int) as age"), data.salary)
```

```python
data_bonus = data_filtered.withColumn('bonus', col("salary")*1.2)
data_bonus.show()
```

```python
data_colrename = data_bonus.withColumnRenamed('bonus', "Salary_Bonus")
data_colrename.show()
```

```python
data_colrename = data_colrename.withColumnRenamed('employee_id', "Emp_id")
data_colrename.show()
```

```python
from pyspark.sql.functions import count
data_groupby = data_colrename.groupby("age").agg(count("age")).alias("Count")

data_groupby.show()
```
