```python
import pandas as pd
import numpy as np

np.random.seed(42)

# -----------------------------
# 1️⃣ Customers Table (30 customers)
# -----------------------------
num_customers = 30

customers = pd.DataFrame({
    "customer_id": range(1001, 1001 + num_customers),
    "customer_name": [f"Customer_{i}" for i in range(1, num_customers + 1)],
    "city": np.random.choice(
        ["Mumbai", "Delhi", "Chennai", "Bangalore", "Hyderabad"],
        num_customers
    ),
    "signup_date": pd.to_datetime("2023-01-01") + 
                   pd.to_timedelta(np.random.randint(0, 180, num_customers), unit="D")
})

# -----------------------------
# 2️⃣ Orders Table (100 rows)
# -----------------------------
num_orders = 100

orders = pd.DataFrame({
    "order_id": range(5001, 5001 + num_orders),
    "customer_id": np.random.choice(customers["customer_id"], num_orders),
    "order_date": pd.to_datetime("2023-06-01") +
                  pd.to_timedelta(np.random.randint(0, 90, num_orders), unit="D"),
    "product_category": np.random.choice(
        ["Electronics", "Clothing", "Groceries", "Home Decor", "Sports"],
        num_orders
    ),
    "amount": np.random.randint(500, 50000, num_orders)
})

# -----------------------------
# 3️⃣ Payments Table (90 rows → some missing payments)
# -----------------------------
num_payments = 90

payments = pd.DataFrame({
    "payment_id": range(8001, 8001 + num_payments),
    "order_id": np.random.choice(orders["order_id"], num_payments, replace=False),
    "payment_method": np.random.choice(
        ["UPI", "Card", "Cash", "Net Banking"],
        num_payments
    ),
    "payment_status": np.random.choice(
        ["Completed", "Failed", "Pending"],
        num_payments,
        p=[0.75, 0.15, 0.10]
    )
})

# Display shapes
print("Customers:", customers.shape)
print("Orders:", orders.shape)
print("Payments:", payments.shape)

customers.head(), orders.head(), payments.head()
```
