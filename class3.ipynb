Problem: Determine account status based on balance: - Balance < 0 → Overdraft - Balance 0–5,000 → Low
Balance - Balance > 5,000 → Normal

```python
balance = [3000,40000,5000,-200,30000,200000,40000]
for bal in balance:
    if bal<=0:
        print("Overdraft")
    elif(bal>0) and (bal<5000):
        print("Low Balance")
    else:
        print("Sufficient Balance!!!")
```

```python
def bal_check(bal):
    if bal<=0:
        print("Overdraft")
    elif(bal>0) and (bal<=5000):
        print("Insufficient funds")
    else:
        Print("Normal")
```

```python
bal_check(5000)
```

Apply discounts based on bill amount: - ≥ 5,000 → 5% - ≥ 10,000 → 10%

```python
amount = int(input("Enter Bill Amount"))
if amount<5000:
    print("Not elligible for discount!!")

elif (amount>=5000) and (amount<=9999):
    final_amount = amount - (amount * 0.05)
    print(f"Your final amount after discount is {final_amount}")
else:
    final_amount = amount - (amount * 0.1)
    print(f"Your final amount after discount is {final_amount}")
```

sales = [20,24,25,0,12,13,15,18,0,11]

Ignore zero-sales days and calculate total, average, and best sale day.

```python
sales = [20,24,25,0,12,13,15,18,0,11]
valid_sales = [i for i in sales if i>0]

print("The total sales is", sum(valid_sales))
print("The average sales is", sum(valid_sales)/len(valid_sales))
print("Top Sales count is", max(valid_sales))
```

Analyze expenses and flag above-average spending.

```python
def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)
```

```python
fact(10)
```
