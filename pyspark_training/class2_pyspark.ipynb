##  join()


```python
from pyspark.sql import Row

# Create sample DataFrame 1 (df1)
data1 = [
    Row(id=1, name='Alice'),
    Row(id=2, name='Bob'),
    Row(id=3, name='Charlie'),
    Row(id=5, name='Eve') # id=5 not in df2 for demonstrating joins
]
df1 = spark.createDataFrame(data1)
df1.show()

# Create sample DataFrame 2 (df2)
data2 = [
    Row(id=1, city='New York'),
    Row(id=2, city='Los Angeles'),
    Row(id=4, city='Chicago'), # id=4 not in df1 for demonstrating joins
    Row(id=3, city='Houston')
]
df2 = spark.createDataFrame(data2)
df2.show()

print("Two sample PySpark DataFrames (df1 and df2) have been created.")
```

```python
print("### Inner Join ###")
# Inner Join: Returns rows when there is a match in both DataFrames.
# Only rows with matching 'id' values in both df1 and df2 will be included.
# Columns from both DataFrames will be present in the result.
inner_join_df = df1.join(df2, on='id', how='inner')
inner_join_df.show()
```

```python
print("\n### Left Outer Join ###")
# Left Outer Join: Returns all rows from the left DataFrame (df1) and the matching rows from the right DataFrame (df2).
# If there is no match, it fills with nulls on the right side. All columns from both DataFrames are included.
left_outer_join_df = df1.join(df2, on='id', how='left_outer')
left_outer_join_df.show()
```

```python
print("\n### Right Outer Join ###")
# Right Outer Join: Returns all rows from the right DataFrame (df2) and the matching rows from the left DataFrame (df1).
# If there is no match, it fills with nulls on the left side. All columns from both DataFrames are included.
right_outer_join_df = df1.join(df2, on='id', how='right_outer')
right_outer_join_df.show()
```

```python
print("\n### Full Outer Join ###")
# Full Outer Join: Returns all rows when there is a match in either DataFrame.
# If there is no match in one of the DataFrames, it fills with nulls on that side.
# All columns from both DataFrames are included.
full_outer_join_df = df1.join(df2, on='id', how='full_outer')
full_outer_join_df.show()
```

```python
print("\n### Left Semi Join ###")
# Left Semi Join: Returns all rows from the left DataFrame (df1) where there is a match in the right DataFrame (df2).
# It's similar to an inner join but only returns columns from the left DataFrame.
left_semi_join_df = df1.join(df2, on='id', how='left_semi')
left_semi_join_df.show()
```

```python
print("\n### Left Anti Join ###")
# Left Anti Join: Returns all rows from the left DataFrame (df1) where there is NO match in the right DataFrame (df2).
# It's used to find rows in df1 that do not have a corresponding entry in df2. Only columns from the left DataFrame are included.
left_anti_join_df = df1.join(df2, on='id', how='left_anti')
left_anti_join_df.show()
```

### `union()`

The `union()` method in PySpark is used to combine two DataFrames with identical schemas (same number of columns, same column names, and compatible data types) into a single DataFrame. It performs a vertical concatenation of the DataFrames. By default, `union()` includes all rows from both DataFrames, **including duplicate rows**.

In newer versions of PySpark (Spark 2.0+), `union()` effectively behaves like `unionAll()` (which is now deprecated) in that it does not remove duplicates automatically. If you need to remove duplicates after a union operation, you must explicitly call the `.distinct()` transformation on the resulting DataFrame.

```python
print("### Union() Demonstration ###")
# union() combines two DataFrames by position, allowing duplicate rows.
# Note: In PySpark 2.0+, union() is equivalent to unionAll().
union_df = df_union1.union(df_union2)
union_df.show()

print("As shown, union() includes duplicate rows (id=1, value='A' and id=2, value='B' appear twice).")
```

### `unionAll()`

The `unionAll()` method in PySpark was historically used to combine two DataFrames with identical schemas, similar to `union()`. It performed a vertical concatenation of the DataFrames by position and **did not remove duplicate rows**.

**Important Note**: In PySpark 2.0 and later, `unionAll()` has been deprecated and its functionality has been merged into the `union()` method. Therefore, `df1.unionAll(df2)` is functionally equivalent to `df1.union(df2)` in modern PySpark, both of which retain duplicates by default. While `unionAll()` might still be found in older codebases, it is recommended to use `union()` for future development.

**Reasoning**:
Now that the deprecation of `unionAll()` has been explained, I will explicitly state that `unionAll()` is no longer available and demonstrate that the `union()` operation, which was performed earlier, effectively represents the historical behavior of `unionAll()` as per the subtask.

### `unionByName()`

The `unionByName()` method in PySpark is a powerful operation used to combine two DataFrames that have compatible schemas but potentially different column orders. Unlike `union()` (which combines DataFrames based on the positional order of their columns), `unionByName()` matches columns by their names. This makes it particularly useful when you are working with DataFrames that originate from different sources or have undergone transformations that alter the column sequence.

**Key Features and Utility:**
*   **Column Name Matching:** It concatenates rows from two DataFrames by matching columns based on their names. This means that if `df1` has `['A', 'B', 'C']` and `df2` has `['C', 'A', 'B']`, `unionByName()` will correctly align `A` with `A`, `B` with `B`, and `C` with `C`.
*   **Order Independence:** It removes the need to manually reorder columns before performing a union operation, which is a common requirement when using `union()`.
*   **Schema Compatibility:** Both DataFrames must have the same column names and compatible data types for those columns. If a column exists in one DataFrame but not the other, `unionByName()` can handle this if `allowMissingColumns` is set to `True`, filling the missing column with nulls in the DataFrame where it's absent.
*   **Duplicate Rows:** Similar to `union()`, `unionByName()` retains duplicate rows by default. To get distinct rows, a subsequent `.distinct()` transformation is required.

`unionByName()` simplifies data integration by providing a flexible way to combine DataFrames without strict reliance on column order, thus reducing boilerplate code and potential errors.

```python
from pyspark.sql import Row

# Create sample DataFrame 1 (df_union_by_name1)
data_ubn1 = [
    Row(id=1, value='Apple', category='Fruit'),
    Row(id=2, value='Banana', category='Fruit'),
    Row(id=3, value='Carrot', category='Vegetable') # Unique to df_union_by_name1
]
df_union_by_name1 = spark.createDataFrame(data_ubn1)
df_union_by_name1.show()

# Create sample DataFrame 2 (df_union_by_name2) with columns in a different order
data_ubn2 = [
    Row(category='Fruit', id=1, value='Apple'), # Common row with df_union_by_name1
    Row(category='Fruit', id=2, value='Banana'), # Common row with df_union_by_name1
    Row(category='Snack', id=4, value='Chips') # Unique to df_union_by_name2
]
df_union_by_name2 = spark.createDataFrame(data_ubn2)
df_union_by_name2.show()

print("Two sample PySpark DataFrames (df_union_by_name1 and df_union_by_name2) have been created with different column orders.")
```

```python
print("### unionByName() Demonstration ###")
# unionByName() combines DataFrames by matching column names, ignoring column order.
# It retains duplicates by default.
union_by_name_df = df_union_by_name1.unionByName(df_union_by_name2)
union_by_name_df.show()

print("As shown, unionByName() correctly combined the DataFrames by column name, despite different column orders.")
print("It also includes duplicate rows (id=1, value='Apple', category='Fruit' and id=2, value='Banana', category='Fruit' appear twice).")
```

### UDF (User Defined Function)

PySpark User Defined Functions (UDFs) allow you to extend PySpark's DataFrame API by defining custom functions in Python that can be applied to columns of a DataFrame. While PySpark offers a rich set of built-in functions, there are often scenarios where you need to implement complex or highly specific logic that isn't available out-of-the-box. UDFs provide the flexibility to bridge this gap.

**Purpose and General Usage:**
*   **Custom Logic:** UDFs are ideal for applying custom row-level transformations to DataFrame columns. For example, formatting strings in a specific way, performing complex mathematical calculations, or implementing business rules.
*   **Integration with Python Ecosystem:** They allow you to leverage the vast Python ecosystem (e.g., NumPy, custom libraries) within your Spark DataFrame operations.
*   **Readability:** Encapsulating complex logic into a function can improve code readability and maintainability.

**How UDFs Work:**
When a UDF is applied to a DataFrame, Spark serializes the Python function and its dependencies, distributes it to worker nodes, and executes it row-by-row on the data partitions. This process involves communication between the JVM (where Spark operates) and Python interpreter (where the UDF runs), which can have performance implications.

**Performance Considerations:**
While powerful, UDFs written in Python can sometimes be a performance bottleneck compared to native PySpark functions or SQL expressions. This is due to:
1.  **Serialization/Deserialization Overhead:** Data needs to be serialized from the JVM to Python and deserialized back for each row processed by the UDF.
2.  **Context Switching:** There's an overhead of switching between the JVM and Python processes.
3.  **Lack of Optimization:** Spark's Catalyst Optimizer cannot optimize operations within a Python UDF, treating it as a black box. This means UDFs don't benefit from Spark's query optimizations like predicate pushdown or columnar processing.

**Best Practices for Performance:**
*   **Prefer Built-in Functions:** Always try to use PySpark's native functions (from `pyspark.sql.functions`) or SQL expressions first, as they are highly optimized.
*   **Vectorized UDFs (Pandas UDFs):** For improved performance, especially with large datasets, consider using Pandas UDFs (also known as Vectorized UDFs). These UDFs execute a Python function on a Pandas Series or DataFrame, allowing Spark to process data in batches rather than row by row, significantly reducing serialization and deserialization overhead.
*   **Minimize UDF Usage:** Use UDFs only when absolutely necessary and when built-in functions cannot achieve the desired logic.



```python
from pyspark.sql import Row

# Create a sample DataFrame for UDF demonstration
data_udf = [
    Row(name='Alice', age=30),
    Row(name='Bob', age=24),
    Row(name='Charlie', age=35),
    Row(name='David', age=29)
]
df_udf = spark.createDataFrame(data_udf)
df_udf.show()

print("Sample DataFrame 'df_udf' created for UDF demonstration.")
```

```python
def to_uppercase(name):
    """Converts a given string to uppercase."""
    if name is None:
        return None
    return name.upper()

print("Python function 'to_uppercase' defined.")
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

# Convert the Python function to a PySpark UDF
upper_case_udf = udf(to_uppercase, StringType())

print("PySpark UDF 'upper_case_udf' created.")
```

```python
from pyspark.sql.functions import col

# Apply the UDF to the 'name' column and create a new column 'name_uppercase'
df_udf_transformed = df_udf.withColumn("name_uppercase", upper_case_udf(col("name")))
df_udf_transformed.show()

print("UDF successfully applied to 'name' column and new 'name_uppercase' column created.")
```

### `transform()`

The `transform()` method in PySpark DataFrames is a powerful and elegant way to apply a sequence of custom transformations to a DataFrame. It is designed to encapsulate complex data manipulation logic into reusable functions, making your code more modular, readable, and maintainable.

**Purpose and Behavior:**
*   **Encapsulating Logic:** `transform()` allows you to define a function that takes a DataFrame as input and returns a DataFrame. This function can contain any number of PySpark DataFrame operations (e.g., `withColumn`, `filter`, `select`, `drop`, `join`, etc.).
*   **Chaining Operations:** The most significant advantage of `transform()` is its ability to be chained directly with other DataFrame operations. Because it takes a DataFrame and returns a DataFrame, you can seamlessly integrate custom transformation functions into a larger pipeline of operations without breaking the fluent API style of PySpark.
*   **Returns New DataFrame:** Crucially, `transform()` always returns a *new* DataFrame. It does not modify the original DataFrame in place, adhering to the immutability principle of DataFrames in Spark. This ensures that your original data remains untouched.

**Analogy:**
Think of `transform()` as a "macro" for DataFrame operations. Instead of writing out a series of `withColumn` or `filter` calls repeatedly, you can bundle them into a single Python function and then apply that function using `df.transform(your_function)`.

**Advantages:**
*   **Readability:** Complex transformation pipelines become easier to understand as distinct logical steps are named functions.
*   **Reusability:** Common transformation sequences can be reused across different DataFrames or different parts of your application.
*   **Modularity:** Promotes modular code design by separating transformation logic from the main data flow.
*   **Functional Style:** Fits well with the functional programming paradigm often used in Spark applications.

While you could achieve similar results by just calling your custom function (e.g., `transformed_df = my_transform_func(df)`), `transform()` offers the benefit of being a member method of the DataFrame, allowing for direct chaining, e.g., `df.filter(...).transform(my_transform_func).select(...)`.

**Reasoning**:
To demonstrate the `transform()` method, I will create a sample PySpark DataFrame with columns suitable for applying a sequence of transformations, as per the second instruction of the subtask.



```python
from pyspark.sql import Row

# Create a sample DataFrame for transform() demonstration
data_transform = [
    Row(id=1, product='Laptop', price=1200.0, quantity=2),
    Row(id=2, product='Mouse', price=25.50, quantity=5),
    Row(id=3, product='Keyboard', price=75.0, quantity=3),
    Row(id=4, product='Monitor', price=300.0, quantity=1)
]
df_transform = spark.createDataFrame(data_transform)
df_transform.show()

print("Sample DataFrame 'df_transform' created for transform() demonstration.")
```

```python
from pyspark.sql.functions import col, when

def apply_product_transformations(df):
    """Applies a sequence of transformations to a product DataFrame."""
    # 1. Calculate total price (price * quantity)
    df_with_total_price = df.withColumn("total_price", col("price") * col("quantity"))

    # 2. Categorize products based on price
    df_categorized = df_with_total_price.withColumn("price_category",
        when(col("price") >= 500, "Expensive")
        .when(col("price") >= 100, "Moderate")
        .otherwise("Affordable")
    )

    # 3. Select and reorder some columns for clarity
    final_df = df_categorized.select("id", "product", "price", "quantity", "total_price", "price_category")

    return final_df

print("Python function 'apply_product_transformations' defined.")
```

```python
print("\n### transform() Method Application ###")
# Apply the custom transformation function using the .transform() method
df_transformed_result = df_transform.transform(apply_product_transformations)
df_transformed_result.show()

print("DataFrame successfully transformed using the .transform() method.")
```

###  `apply()`

In PySpark, the concept of `apply()` as found in Pandas DataFrames (for applying a function element-wise or row-wise) does not have a direct, identically named counterpart on the PySpark DataFrame API for general row-wise operations. Instead, PySpark provides several powerful mechanisms to achieve similar functionality, primarily catering to distributed and optimized execution. These methods are broadly categorized into two main approaches:

#### 1. Row-wise application using UDFs with `withColumn`
For applying a custom function to a column (or multiple columns) on a row-by-row basis, the most common approach is to use PySpark's User Defined Functions (UDFs) in conjunction with DataFrame transformations like `withColumn` or `select`.

*   **Mechanism**: You define a standard Python function, register it as a UDF (specifying its return type), and then apply it to a column using `df.withColumn("new_col", my_udf(col("existing_col")) )`.
*   **Purpose**: This is suitable for custom logic that needs to be applied independently to each row's value(s) within a column. Examples include custom string formatting, simple arithmetic not covered by built-in functions, or conditional logic.
*   **Benefits**: Simple to implement for straightforward custom logic. Leverages the Python ecosystem.
*   **Limitations**: Can be a performance bottleneck due to serialization/deserialization overhead and the Python interpreter's overhead (JVM to Python context switching) for each row. Spark's Catalyst Optimizer cannot optimize operations inside Python UDFs.

#### 2. Vectorized application using Pandas UDFs (Arrow-optimized UDFs)
For significantly improved performance, especially with large datasets and complex logic that can be expressed using Pandas operations, PySpark introduced Pandas UDFs (also known as Vectorized UDFs). These leverage Apache Arrow to transfer data efficiently between JVM and Python processes in batches, minimizing serialization overhead.

Pandas UDFs are primarily used in two main scenarios:

##### a) Series-to-Series Pandas UDFs (similar to `apply()` on a single Series)
*   **Mechanism**: These UDFs take a Pandas Series as input and return a Pandas Series of the same length. They are applied using `withColumn` similar to traditional UDFs, but operate on batches of data.
*   **Purpose**: Ideal for column-wise operations where the logic can be applied to a Pandas Series, e.g., complex numerical transformations, string manipulations that benefit from Pandas' vectorized operations.
*   **Benefits**: Much faster than traditional Python UDFs due to vectorized execution and reduced overhead. Allows leveraging the rich Pandas API.

##### b) Grouped Map Pandas UDFs (`groupBy().applyInPandas()`)
*   **Mechanism**: This is the closest equivalent to `groupBy().apply()` in Pandas. The UDF takes a Pandas DataFrame (representing a group of rows) as input and must return a Pandas DataFrame. You specify the output schema for the returned DataFrame.
*   **Purpose**: Enables applying arbitrary Pandas DataFrame transformations to each group of data. This is powerful for tasks like calculating rolling averages, group-specific statistics, time-series analysis within groups, or fitting models per group.
*   **Benefits**: Highly efficient for group-wise operations. Allows full flexibility of the Pandas DataFrame API within each group, while Spark handles the distributed grouping and execution.

##### c) Map Pandas UDFs (`mapInPandas()`)
*   **Mechanism**: This method transforms each partition of a PySpark DataFrame into a Pandas DataFrame, applies a Python function that takes and returns a Pandas DataFrame, and then converts the result back into a PySpark DataFrame.
*   **Purpose**: Useful for highly custom transformations that need to operate on an entire partition's worth of data at once, where `groupBy().applyInPandas()` might not fit (e.g., when you don't have an explicit grouping key but want to process partitions).
*   **Benefits**: Provides maximum flexibility to use Pandas on partitions, often for advanced data preparation or model inference steps.

In summary, while PySpark doesn't have a singular `apply()` method, its UDF and Pandas UDF functionalities, especially `groupBy().applyInPandas()`, provide robust and performant ways to apply custom Python logic to DataFrames, ranging from element-wise to group-wise operations.

```python
from pyspark.sql import Row

# Create a sample DataFrame for apply() demonstration
data_apply = [
    Row(id=1, category='A', value=10, timestamp='2023-01-01 10:00:00'),
    Row(id=2, category='A', value=20, timestamp='2023-01-01 10:05:00'),
    Row(id=3, category='B', value=15, timestamp='2023-01-01 10:10:00'),
    Row(id=4, category='B', value=25, timestamp='2023-01-01 10:15:00'),
    Row(id=5, category='A', value=30, timestamp='2023-01-01 10:20:00'),
    Row(id=6, category='C', value=5, timestamp='2023-01-01 10:25:00')
]
df_apply = spark.createDataFrame(data_apply)
df_apply.show()

print("Sample DataFrame 'df_apply' created for apply() demonstration.")
```

```python
from pyspark.sql.functions import udf, col
from pyspark.sql.types import IntegerType

def multiply_by_factor(value, factor=2):
    """Multiplies a value by a given factor."""
    if value is None:
        return None
    return value * factor

# Register the Python function as a PySpark UDF
# We expect the output to be an Integer, so we specify IntegerType
multiply_udf = udf(multiply_by_factor, IntegerType())

print("### Row-wise UDF Application with withColumn ###")
# Apply the UDF to the 'value' column to create a new column 'multiplied_value'
df_with_udf = df_apply.withColumn("multiplied_value", multiply_udf(col("value")))
df_with_udf.show()

print("UDF 'multiply_by_factor' applied row-wise to 'value' column.")
```

```python
from pyspark.sql.types import StructType, StructField, StringType, TimestampType, DoubleType
import pandas as pd

output_schema = StructType([
    StructField("category", StringType()),
    StructField("timestamp", TimestampType()),
    StructField("metric_value", DoubleType())
])

def calculate_group_metrics(pdf):
    # Example: compute mean of some column
    metric_value = pdf['value'].mean()
    # Ensure proper datetime type
    timestamp = pd.to_datetime('now')

    return pd.DataFrame({
        'category': [pdf['category'].iloc[0]],
        'timestamp': [timestamp],
        'metric_value': [metric_value]
    })

df_grouped_pandas_udf = df_apply.groupBy("category").applyInPandas(calculate_group_metrics, schema=output_schema)
df_grouped_pandas_udf.show()
```

```python
import pandas as pd
from pyspark.sql.functions import pandas_udf, PandasUDFType
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType, TimestampType

# Define the output schema for the Grouped Map Pandas UDF
# The output DataFrame must conform to this schema.
output_schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("category", StringType(), True),
    StructField("value", IntegerType(), True),
    StructField("timestamp", TimestampType(), True),
    StructField("normalized_value", DoubleType(), True),
    StructField("rolling_mean_3d", DoubleType(), True)
])

# Define a Python function that will operate on a Pandas DataFrame for each group
def calculate_group_metrics(pdf: pd.DataFrame) -> pd.DataFrame:
    # Convert timestamp to datetime objects within the pandas DataFrame for sorting and schema compatibility
    pdf['timestamp'] = pd.to_datetime(pdf['timestamp'])

    # Sort by timestamp within each group for rolling calculations
    pdf = pdf.sort_values(by='timestamp')

    # Calculate normalized value (value / sum of values in the group)
    sum_value = pdf['value'].sum()
    if sum_value > 0: # Avoid division by zero
        pdf['normalized_value'] = pdf['value'] / sum_value
    else:
        pdf['normalized_value'] = 0.0

    # Calculate a 3-period rolling mean for 'value'
    # min_periods=1 means it will start calculating with the first value
    pdf['rolling_mean_3d'] = pdf['value'].rolling(window=3, min_periods=1).mean()

    return pdf

print("Python function 'calculate_group_metrics' and output schema defined for Grouped Map Pandas UDF.")
```

```python
print("\n### Grouped Map Pandas UDF Application with applyInPandas() ###")
# Apply the Pandas UDF to grouped data. The output schema is mandatory.
# We group by 'category' and apply our custom Pandas function to each group.
df_grouped_pandas_udf = df_apply.groupBy("category").applyInPandas(calculate_group_metrics, schema=output_schema)
df_grouped_pandas_udf.show()

print("Grouped Map Pandas UDF 'calculate_group_metrics' applied using applyInPandas().")
```
