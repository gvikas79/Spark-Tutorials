```python
emp.show()
```

```python
from pyspark.sql.functions import col
emp = emp.withColumn("age", col("age").cast("int")).withColumn("salary", col("salary").cast("int")).withColumn("hire_date", col("hire_date").cast("date"))
```

```python
emp.printSchema()
```

```python
emp.filter((col("name").startswith("J")) & (col("salary")>=50000)).show()
```

```python
emp.filter(col("department_id").isin("101","102")).show()
```

```python
emp.groupBy("department_id").count().show()
```

```python
emp.groupBy("department_id").avg("salary").show()
```

```python
from pyspark.sql.functions import avg

emp.groupBy(["department_id", "gender"]).agg(avg("salary").alias("total_sal")).show()
```
