```python
emp.show()
```

```python
emp_filtered = emp.where("gender == 'Female'")
emp_filtered_sal = emp_filtered.where("salary >= 25000")
#emp_filtered.show()

emp_filtered_sal.show()
```

```python
emp_bonus = emp.withColumn("bonus", col("salary") * 1.2)
emp_bonus.show()
```

```python
emp.printSchema()
```

```python
from pyspark.sql.functions import to_date
emp_date_fix = emp.withColumn("hire_date", to_date(col("hire_date"), "yyyy-MM-dd"))
emp_date_fix.show()
```

```python
emp_date_fix.printSchema()
```

```python
emp.select(col("gender")).distinct().show()
```

```python
emp.filter((col("salary")>50000) & (col("age")>30)).show(5)
```

```python
emp.orderBy(col("salary").desc()).show()
```

```python
from pyspark.sql.functions import year, month, dayofweek, dayname

emp_date_fix.withColumn("year", year("hire_date")).show(2)
emp_date_fix.withColumn("month", month("hire_date")).show(2)
emp_date_fix.withColumn("day", dayofweek("hire_date")).show(2)
emp_date_fix.withColumn("day", dayname("hire_date")).show(2)
```

```python
from pyspark.sql.functions import when
emp.withColumn("gender",
              when(col("gender") == "Male", "M")
              .when(col("gender") == "Female", "F")
             ).show(5)
```

```python
emp.select(col("gender")).distinct().count()
```

```python
from pyspark.sql.functions import avg, sum, min, max

emp.select(avg("salary")).show()
emp.select(sum("salary")).show()
emp.select(max("salary")).show()
emp.select(min("salary")).show()
```
