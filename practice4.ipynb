```python
import pandas as pd
import numpy as np

np.random.seed(42)

products = pd.DataFrame({
    "product_id": range(1001, 1021),
    "product_name": [f"Product_{i}" for i in range(1, 21)],
    "category": np.random.choice(
        ["Electronics", "Clothing", "Home", "Sports"],
        20
    ),
    "unit_cost": np.random.randint(200, 2000, 20),
    "unit_price": np.random.randint(500, 5000, 20),
    "reorder_level": np.random.randint(20, 100, 20)
})

print(products.head())

warehouses = pd.DataFrame({
    "warehouse_id": [1, 2, 3],
    "warehouse_city": ["Mumbai", "Delhi", "Bangalore"]
})

print(warehouses.head())

inventory = pd.DataFrame({
    "warehouse_id": np.random.choice(warehouses["warehouse_id"], 60),
    "product_id": np.random.choice(products["product_id"], 60),
    "stock_quantity": np.random.randint(0, 300, 60),
    "last_updated": pd.to_datetime("2023-07-01") +
                    pd.to_timedelta(np.random.randint(0, 30, 60), unit="D")
})

print(inventory.head())

sales = pd.DataFrame({
    "sale_id": range(5001, 5151),
    "product_id": np.random.choice(products["product_id"], 150),
    "warehouse_id": np.random.choice(warehouses["warehouse_id"], 150),
    "sale_date": pd.to_datetime("2023-06-01") +
                 pd.to_timedelta(np.random.randint(0, 60, 150), unit="D"),
    "quantity_sold": np.random.randint(1, 20, 150)
})

print(sales.head())

purchases = pd.DataFrame({
    "purchase_id": range(8001, 8051),
    "product_id": np.random.choice(products["product_id"], 50),
    "warehouse_id": np.random.choice(warehouses["warehouse_id"], 50),
    "purchase_date": pd.to_datetime("2023-05-01") +
                     pd.to_timedelta(np.random.randint(0, 60, 50), unit="D"),
    "quantity_purchased": np.random.randint(10, 100, 50)
})

print(purchases.head())
```

What is total revenue generated from sales?

🔹 GROUPBY PRACTICE

Total sales quantity per product.

Total revenue per warehouse.

Average selling quantity per product category.

Monthly sales trend.

Total purchase quantity per product.

🔹 JOIN PRACTICE

Join sales with products to calculate revenue.

Join inventory with products to find stock value.

Join inventory with warehouses.

Find products where stock_quantity < reorder_level.

Find warehouse-wise low stock products.

🔹 INTERMEDIATE ANALYSIS

Calculate profit per product:
(unit_price - unit_cost) × total quantity sold

Identify top 5 best-selling products.

Find slow-moving products (lowest sales).

Find dead stock:

Products with stock > 100 but sales < 10

Calculate inventory turnover ratio:
(Total Sales / Average Inventory)

🔹 ADVANCED INVENTORY ANALYTICS

Calculate stock aging:

Days since last_updated

Identify products not sold in last 30 days.

Find overstocked products:

High stock + low sales

Identify stock-out risk:

Sales trend high but stock low

Forecast demand using last 7-day moving average.

🔥 HARD / INTERVIEW LEVEL

Simulate:

If sales increase by 20%, which products will hit stock-out?

Create a replenishment plan:

Suggest purchase quantity = reorder_level - current stock

| Warehouse | Total Stock | Total Sales | Revenue | Low Stock Count |

ABC Analysis:

Classify products into A/B/C based on revenue contribution.

💣 ULTRA HARD (Supply Chain Analyst Level)

Detect anomaly sales:

quantity_sold > mean + 2*std

Calculate service level:
